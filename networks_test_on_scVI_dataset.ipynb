```python
#This is the Dataset class. It combined with Dataloader can be helpful in training
from torchvision import transforms#, utils
from torch.utils.data import Dataset, DataLoader
#class ScVIDataset can accept transforms that can transform the data
class ScVIDataset(Dataset):
    '''scVI dataset, 
        it will output {'features':features, 'target':target}
        it will take an optional argument 'transform' so that any required processing can be applied on the sample.
    '''
    def __init__(self, data_frame, target_col='cell_labels', transform=None):
        '''
        Args:
            data_frame(pandas.core.frame.DataFrame): It contains the dataframe
            target_col(string, default is 'cell_labels'): the target label column's name. default is 'cell_labels'
            transform(callable, optional): The transform that can be applied on a sample
        '''
        self.data_frame = data_frame.drop([target_col],axis=1)
        self.target = data_frame[target_col]
        self.transform = transform
                
    def __len__(self):
        return len(self.data_frame)
    
    def __getitem__(self, idx):
        #target = np.zeros(len(self.target_map))##DEL
        #target[self.target_map[self.target.iloc[idx]]] = 1##DEL
        sample = {'features':  self.data_frame.iloc[idx], 'target':self.target.iloc[idx].astype('int64')}
        if self.transform:
            sample = self.transform(sample)
        
        return sample
```

```python
#define the transforms we should use here:
#Normalize: to normalize to given mean and std
#ToTensor: to convert the pandas data frame to torch tensors
class Normalize(object):
    '''
    This is (feature - mean)/std for each feature in features
    Args:
        means(numpy.ndarray or python list): a list of means
        stds(numpy.ndarray or python list): a list of stds
    '''
    
    def __init__(self, means, stds):
        self.means = means
        self.stds = stds
    
    def __call__(self, sample):
        return {'features': (sample['features']-self.means)/self.stds, 'target': sample['target']}
    
class ToTensor(object):
    '''Convert the data frames in sample to Tensors'''
    def __call__(self, sample):
        #torch image: C X H X W
        feature = sample['features'].values # the shape is (#,), So, I need to expand its dimension
        return {'features': torch.from_numpy(feature[None,None,:]), 
            'target': sample['target']}#torch.from_numpy(sample['target'])}

```

```python
#Prepare the train_loader and valid_loader
train_transforms = transforms.Compose([Normalize(norm_mean, norm_std),
                                       ToTensor()])

valid_transforms = transforms.Compose([Normalize(norm_mean, norm_std),
                                       ToTensor()])

train_datasets = ScVIDataset(data_train, target_col='cell_labels', transform = train_transforms)
valid_datasets = ScVIDataset(data_valid, target_col='cell_labels', transform = valid_transforms)

batch_size = 20
num_workers = 0
train_loader = DataLoader(train_datasets, batch_size=batch_size, shuffle=True, num_workers=num_workers)
valid_loader = DataLoader(valid_datasets, batch_size=batch_size, shuffle=True, num_workers=num_workers)
data_loaders = {'train':train_loader, 'valid':valid_loader}
```

```python
#implementation of the fcmodel
import torch.nn as nn
import torch.nn.functional as F

#define the architecture of fcmodel: FCNet
class FCNet(nn.Module):
    def __init__(self):
        super(FCNet, self).__init__()
        #define the layers in this model
        #input layer is size len(features)=558
        self.fc1 = nn.Linear(len(features), 256)
        self.fc2 = nn.Linear(256,256)
        self.fc3 = nn.Linear(256,10)
        self.fc4 = nn.Linear(10,len(labels))
        
        self.dropout = nn.Dropout(0.5)
        self.relu = nn.ReLU(inplace=True)
        
    def forward(self, x):
        #Define the forward behavior
        #flat input data first
        x = x.view(x.shape[0], -1)
        x = self.relu(self.fc1(x))
        x = self.dropout(self.relu(self.fc2(x)))
        x = self.dropout(self.relu(self.fc3(x)))
        x = self.fc4(x)
        
        return x        
```

```python
# the training function
def train(epochs, loaders, model, optimizer, criterion, use_cuda, save_path):
    '''Training function
    Args:
        epochs: the total number of training epochs
        loaders(python dict):with key='train' and 'valid'
        model(): training model
        optimizer:
        criterion:
        use_cuda(boolean): True if cuda is available
        save_path(string of path): the parameters will be saved to the path after training
    '''
    
    valid_loss_min = np.Inf
    
    for epoch in range(1, epochs+1):
        #Initialized to monitor training loss and validatoin loss during training
        train_loss = 0
        valid_loss = 0
        
        # train loop
        #set model to train mode
        model.train()
        for batch_idx, sample in enumerate(loaders['train']):
            data, target = sample['features'], sample['target']
            if use_cuda:
                data, target = data.cuda(), target.cuda()
            optimizer.zero_grad()
            output = model(data)
           
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            
            #The loss here is averaged by batch, 
            #  So, using the following equation to calculate the average training_loss during training
            train_loss = train_loss + ((1 / (batch_idx + 1)) * (loss.data - train_loss))
        
        
        # validation loop
        #set the model to evaluation mode
        model.eval()
        #disable auto_grad to speed things up in validation
        with torch.no_grad():
            for batch_idx, sample in enumerate(loaders['valid']):
                data, target = sample['features'], sample['target']
                if use_cuda:
                    data, target = data.cuda(), target.cuda()
                
                output = model(data)
                loss = criterion(output, target)
                valid_loss=valid_loss + ((1 / (batch_idx + 1)) * (loss.data - valid_loss))
                
        #print train loss and valid loss for each epoch
        print(f'Epoch:{epoch} \tTraining Loss:{train_loss:.6f} \tValidation Loss:{valid_loss:.6f}' )
        
        #save the model parameters if validation loss has decreased
        if valid_loss <= valid_loss_min:
            print(f'--Validation loss decreased({valid_loss_min:.6f} --> {valid_loss:.6f})')
            torch.save(model.state_dict(), save_path)
            valid_loss_min = valid_loss
        
    
    return model
```

```python
#instantiate the model
model_FCNet = FCNet()

is_cuda_available = torch.cuda.is_available()
if is_cuda_available:
    model_FCNet.cuda()
```

```python
#the test function that test the accuracy of the model
def test(test_loader, model, criterion, use_cuda):
    '''Test function
    Args:        
        loaders(python dict):with key='train' and 'valid'
        model(): training model        
        criterion:
        use_cuda(boolean): True if cuda is available        
    '''
    test_loss = 0.0
    correct = 0.0
    total = 0.0
        
    model.eval()
    for batch_idx, sample in enumerate(test_loader):
        data, target = sample['features'], sample['target']
        if use_cuda:
            data, target = data.cuda, target.cuda
        
        output = model(data)
        loss = criterion(output, target)
        test_loss = test_loss + (1/(batch_idx+1)*(loss.data - test_loss))
        
        #compute the probability using softmax
        prob = F.softmax(output, dim=1)
        #find the prediction from the max prob
        pred = prob.data.max(1, keepdim=True)[1]
        #compare predictions and target 
        correct += np.sum(np.squeeze(pred.eq(target.data.view_as(pred))).cpu().numpy())
        total += data.shape[0]
    
    print(f'Loss: {test_loss:.6f}')
    print(f'Accuracy: {100. * correct / total:.2f}%, ({correct}, {total})')
    
```
